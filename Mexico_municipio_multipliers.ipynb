```python
import pandas as pd

import matplotlib.pyplot as plt

import matplotlib as mpl

import numpy as np

%matplotlib inline

plt.style.use('ggplot')

pd.set_option('precision',3)

from IPython.display import display, HTML

pd.__version__
```

```python
CoastalID = pd.read_csv("CoastalID.csv")
Production = pd.read_csv("ProduccionBrutaTotal.csv")
AddedValue = pd.read_csv("ValorAgregadoBruto.csv")
Municipios = pd.read_csv("Coastal_municipio.csv")
```

```python
Production.columns.values
```

```python
AddedValue.columns.values
```

```python
Municipios.columns.values
```

```python


Production_Value = pd.merge(
    left=Production, 
    right=AddedValue, 
    how='left'
)
```

```python

Product_Value = pd.merge(
    left=Production_Value, 
    right=CoastalID, 
    how='left'
)
```

```python
Product_Value['EconomicImpact'] = Product_Value['TotalProduction']+Product_Value['AddedValue']  
```

```python
Product_Value['Multiplier'] = Product_Value['EconomicImpact'] / Product_Value['TotalProduction'] 
```

```python
Product_Value.columns.values
```

```python
Product_Value = pd.merge(
    left=Product_Value, 
    right=Municipios, 
    how='left'
)
```
